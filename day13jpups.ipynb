```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

x1 = np.linspace(0 ,10, 100)

#plot figure
fig = plt.figure()

plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
```

```python
plt.figure()

#create first of two panel and set current sin plot
plt.subplot(2,1,1) #(rows,columns,panel number)
plt.plot(x1,np.sin(x1),'-')

plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1),'--')
```

```python
plt.figure()

plt.subplot(2,2,1) 
plt.plot(x1,np.sin(x1),'-')

plt.subplot(2,2,2)
plt.plot(x1,np.cos(x1),'--')

plt.subplot(2,2,3) 
plt.plot(x1,np.sin(x1),'-')

plt.subplot(2,2,4)
plt.plot(x1,np.cos(x1),'--')
```

```python
# get current figure info
print(plt.gcf())
```

```python
# get current axis info
print(plt.gca())
```

```python
plt.figure()

#create first of two panel and set current sin plot
plt.subplot(2,1,1)
#(rows,columns,panel number)
plt.plot(x1,np.sin(x1),'-')

plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1),'--')

plt.xlabel('X axis')
plt.ylabel('Y axis ')

plt.show()
```

```python
x = np.linspace(0,2,100)
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')

plt.xlabel('X axis')
plt.ylabel('Y axis ')

plt.title('Simple Plot')

plt.legend() # helps in better understand of diff function 
plt.show()
```

```python
import seaborn as sns
sns.set_style('darkgrid')
```

```python
tips=sns.load_dataset("tips") # loading tips dataset
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

```python
# create a scatter plot of total bill
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.show()
```

```python
# create a line plot of time vs passenger from the titanic dataset 
sns.lineplot(data=titanic,x='age',y='fare')
plt.show()
```

```python
# create a bar plot of day vs total_bill from the tips dataset 
sns.barplot(data=tips,x='day',y='total_bill')
plt.show()
```

```python
# create a histogram  total_bill from the tips dataset 
sns.histplot(data=tips,x='total_bill')
plt.show()
```

```python
# create a box plot of day vs total_bill from the tips dataset 
sns.boxplot(data=tips,x='day',y='total_bill')
plt.show() # dots are outliers
```

```python
# create heatmap  of the correlayion matrix for the ieis dataset
corr_matrix = iris.corr()
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
# create pairplot of iris dataset 
sns.pairplot(data=iris)
plt.show()
```
