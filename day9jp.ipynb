```python
img1= plt.imread("pic2.jpg")
plt.imshow(img1[:,:,::-1])
```

```python
img_gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
th ,img_bin=cv2.threshold(img_gray,100,255,cv2.THRESH_BINARY)
img_bin.shape
```

```python
plt.imshow(img_bin,cmap='gray')
```

```python
#image cropping
plt.imshow(img1[300:700,400:800,::-1])
```

```python
x,y,w,h=(50,150,100,140)

plt.imshow(img1[y:y+h,x:x+w,::-1])
plt.show()
x1,y1,x2,y2=(50,150,270,230)
plt.imshow(img1[y1:y2,x1:x2,::-1])
plt.show()
y1,x1,y2,x2 = (50,150, 270, 210)
plt.subplot(1,3,3)
plt.imshow(img1[y1:y2, x1:x2, ::-1])
```

```python
# image scaling 
img_large=cv2.resize(img1,(1500,1500),interpolation=cv2.INTER_CUBIC)
plt.imshow(img_large[:,:,::-1])
img_large.shape
```

```python
img_small=cv2.resize(img1,(150,150),interpolation=cv2.INTER_CUBIC)
plt.imshow(img_small[:,:,::-1])
img_small.shape
```

```python
img_small=cv2.resize(img1,(0,0), fx=0.25,fy=0.25,interpolation=cv2.INTER_CUBIC)
plt.imshow(img_small[:,:,::-1])
img_small.shape
```

```python
# image rotate 
img_90=cv2.rotate(img1,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_90[:,:,::-1])
```

```python

rot_mat = cv2.getRotationMatrix2D(
    center=(img1.shape[1]//2,img1.shape[0]//2), 
    angle=-30, scale = 1)
img_x = cv2.warpAffine(
    img1, rot_mat, dsize=(img1.shape[1], img1.shape[0]),
    flags=cv2.INTER_LINEAR
)
plt.imshow(img_x[:,:,::-1]); plt.show()

img_x = ski.transform.rotate(img1, -30)
plt.imshow(img_x[:,:,::-1])
```

```python
plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(img1[:,:,::-1]); plt.axis('off')

plt.subplot(1,3,2)
plt.title('Brightened')
plt.imshow(cv2.add(img1[:,:,::-1], 20)); plt.axis('off')

plt.subplot(1,3,3)
plt.title('Darkened')
plt.imshow(cv2.subtract(img1[:,:,::-1], 20)); plt.axis('off')
```

```python
plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(img1[:,:,::-1]); plt.axis('off')

plt.subplot(1,3,2)
plt.title('Contrast +')
plt.imshow(cv2.multiply(img1[:,:,::-1], 1.1)); plt.axis('off')

plt.subplot(1,3,3)
plt.title('Contrast -')
plt.imshow(cv2.divide(img1[:,:,::-1], 1.5)); plt.axis('off')
```

```python
# alpha -> Contrast,  beta -> Brightness
img_cntbrt = cv2.convertScaleAbs(
    img1, alpha = 1.25, beta = 0
)
plt.imshow(img_cntbrt[:,:,::-1])
```
